machine learning with python : save and load trained model

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df_home_price=pd.read_csv("/content/py/ML/4_save_model/homeprices.csv")
```

```python
df_home_price
```

```python
model_to_predict=linear_model.LinearRegression()
model_to_predict.fit(df_home_price[['area']],df_home_price[['price']])
```

```python
model_to_predict.coef_
```

```python
model_to_predict.intercept_
```

```python
df_home_price
```

```python
model_to_predict.predict([[3000]])
```

```python
import pickle as pk
```

```python
with open('model_pk','wb') as file :
  pk.dump(model_to_predict,file)
```

```python
with open('model_pk','rb') as file :
  my_predictor_model=pk.load(file)
```

```python
my_predictor_model.predict([[300]])
```

```python
my_predictor_model.predict([[5000]])
```

```python
from statistics import mode
```

```python
df_airlines.drop('Aircraft',axis=1,inplace=True)
```

```python
df_airlines
```

```python
df_airlines['AirlineName'].fillna(mode(df_airlines['AirlineName']),inplace=True)
```

```python
df_airlines['CabinType'].fillna(mode(df_airlines['CabinType']),inplace=True)
```

```python
df_airlines.drop('DateFlown',axis=1,inplace=True)
```

```python
df_airlines
```

```python
df_airlines['EntertainmentRating'].fillna(mode(df_airlines['EntertainmentRating']),inplace=True)
```

```python
df_airlines['OriginCountry'].unique()
```

```python
df_airlines.drop('Route',axis=1,inplace=True)
```
